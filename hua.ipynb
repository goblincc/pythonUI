```python
import datetime
def string2Month(strs):
    date = datetime.datetime.strptime(strs,'%Y-%m-%d %H:%M:%S')
    month = str(int(date.strftime("%m")))
    return month
```

```python
def string2Year(strs):
    date = datetime.datetime.strptime(strs,'%Y-%m-%d %H:%M:%S')
    year = date.strftime("%Y")
    return year

str(int(string2Year("2021-02-01 00:00:00")) + 1)
```
